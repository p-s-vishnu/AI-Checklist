### Line Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# %matplotlib inline
```

```python
x1 = np.array(range(2,9,2))
y1 = x1**2
y1
```

```python
x2 = x1
y2 = x2 ** 3
y2
```

```python
plt.plot(x1,y1, linestyle="solid",label="plot1",marker="o",color="r")
plt.plot(x2,y2,linestyle="dotted",label="plot2",marker="v",color="g")

plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")

# Custom a and y indexes 
plt.xticks([2,4,6,8,10])
# plt.yticks([12,24,36,48,90])

plt.title("First Plot")
plt.legend(fontsize=18,loc=0)
plt.savefig("myplot.png")
plt.show()
print()
```

### Scatter Plot

```python
x1,y1

plt.scatter(x1,y1,label="scatter points",marker="*",color="b")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

## Bar Chart

```python
stock_names = ["GOOG", "APPL", "IBM", "INTC", "CSCO"]
stock_value = [1000,800,200,400,800]

x_pos = range(len(stock_names))

plt.bar(x_pos,stock_value,width=0.35, label="ticker",align="center",color="b",alpha=0.3)
plt.xticks(x_pos, stock_names)
plt.xlabel("Companies")
plt.ylabel("Price")
plt.title("Current stock values")
plt.legend()
plt.show()
```

```python
sc = df['SaleCondition'].value_counts()
# type(sc)

sc.plot(kind='bar',label="count",color='r',alpha=0.5)
plt.title('Sales of various house types')
plt.legend()
plt.show()
```

### Heat Map

```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(12,9))
# sns.heatmap(corrmat,square=True,vmax=0.8)
sns.heatmap(corrmat)
plt.show()
```

### Distribution Plot (Histogram + Kernel Density Estimate point)

```python
sns.distplot(df['SalePrice'],bins=10,kde=True);
```

### Rug plot

```python
f, ax = plt.subplots(figsize=(12,9))
sns.distplot(df['SalePrice'],bins=50,kde=False,rug=True);
```

### Pair plot

```python
sns.set()
cols = ['SalePrice','OverallQual','GrLivArea','GarageCars','TotalBsmtSF','FullBath']
sns.pairplot(df[cols],size=3)
plt.show()
```

### Swarm Plot

```python
help(sns.axes_style)
```

```python
fig, axis = plt.subplots(figsize=(12,9))
# {darkgrid, whitegrid, dark, white, ticks}
sns.set()
sns.swarmplot(x="SaleCondition",y="SalePrice",data=df)
print(plt.show())
```

```python
df['SalePriceSF'] = df['SalePrice']/df['GrLivArea']
fig, ax = plt.subplots(figsize=(10,8))
plt.hist(df['SalePriceSF'],bins=15,color="gold")
plt.title("Sale price per square foot")
plt.xlabel("Price per spuare feet")
plt.ylabel("Number of Sales")
plt.show()
```

### Joint Plot ( Histogram and Scatter plot)

```python
sns.jointplot(df['LotFrontage'],df['LotArea'],color='grey',label="lot area per frontage")
plt.legend()
plt.xlabel('Lot Frontage')
plt.ylabel('Lot Area')
plt.show()
```
