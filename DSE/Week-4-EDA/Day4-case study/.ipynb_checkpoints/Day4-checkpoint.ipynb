### Revolving Utilization Of Unsecured Lines    

#### Treating outliers

```python
df1["RevolvingUtilizationOfUnsecuredLines"].describe()
```

```python
df1[df1['RevolvingUtilizationOfUnsecuredLines'] > 1].shape
```

There are 3021 observations with revolving utilization of unsecured line greater than 1, since this number is significant we dont want to drop it. Hence, these rows will be converted to missing values and impute using distribution based method.

```python
dfru = df1[df1['RevolvingUtilizationOfUnsecuredLines'] < 1]
# print
ax = sns.distplot(dfru['RevolvingUtilizationOfUnsecuredLines'])
```

#### Replacing higher values with Nan value

```python
df1['RevolvingUtilizationOfUnsecuredLines'] = df1['RevolvingUtilizationOfUnsecuredLines'].map(lambda x: x if x < 1 else np.NaN)
df1['RevolvingUtilizationOfUnsecuredLines'].describe()
```

```python
df1['RevolvingUtilizationOfUnsecuredLines'].fillna(method="ffill",inplace=True)

df1['RevolvingUtilizationOfUnsecuredLines'].describe()
```

```python
ax = sns.distplot(df1['RevolvingUtilizationOfUnsecuredLines'])
```

```python
df1.info()
```

### MonthlyIncome

```python
df1['MonthlyIncome'].describe()
```

```python
sns.distplot(df1['MonthlyIncome'].dropna())
```

```python
sns.boxplot(df1['MonthlyIncome'].dropna())
```

Instead of replacing the outliers we can replace the outliers as missing values. 

Then fill the missing values with a distributed value.

```python
q1 = df1['MonthlyIncome'].quantile(q=0.25)
q3 = df1['MonthlyIncome'].quantile(q=0.75)
print(q1,q3)
```

```python
iqr = q3 - q1
print(iqr)
```

```python
upper_limit = q3 + 1.5*iqr
upper_limit
```

```python
# This is based on finance 
lower_limit = 1000
```

Replacing the values outside the range as NaN

```python
df1['MonthlyIncome'] = df1['MonthlyIncome'].map(lambda x: x if x >= lower_limit and x <= upper_limit else np.NaN)
df1['MonthlyIncome'].describe()
```

```python
df1['MonthlyIncome'].fillna(method="ffill",inplace=True)
df1['MonthlyIncome'].describe()
```

```python
ax = sns.distplot(df1['MonthlyIncome'])
plt.grid(True)
```

```python
df1.info()
```

#### Outlier treatment

```python
df1[(df1['age']<18) | (df1['age']>80)].shape
```

Finance industry expects that the customers are between `18` and `80` years of age. However, according to the data there are 4937 outlier. 

We will follow the previous approach to replace the outliers as np.NaN and fillna with a distribution based imputation.

```python
df1['age'] = df1['age'].apply(lambda x : np.NaN if ((x<18)|(x>80)) else x )
df1['age'].describe()
```

```python
df1['age'].fillna(method="ffill",inplace=True)
df1['age'].describe()
```

```python
ax = sns.distplot(df1['age'])
```

```python
df1.info()
```
