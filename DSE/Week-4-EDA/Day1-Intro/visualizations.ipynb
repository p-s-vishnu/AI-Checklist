## Univariate analysis

```python
ax = sns.barplot(x="sex",y="total_bill",data=tips)
```

As the chart shows male customers (mean=$20.74) are likely to pay more tip compared to female customers.

```python
ax = sns.boxplot(x="time",y="total_bill",data=tips)
```

```python
ax = sns.boxplot(x="sex",y="total_bill",hue="smoker",data=tips)
```

```python
ax = sns.boxplot(x="sex",y="tip",hue="smoker",data=tips)
```

```python
ax = sns.violinplot(x="sex",y="tip",hue="smoker",split=True,data=tips)
```

```python
tips.columns
```

```python
ax = sns.violinplot(x="day",y="total_bill",hue="sex",split=True,data=tips)
```

Females were more conservative while eating on fridays

```python
ax = sns.violinplot(x="time",y="total_bill",hue="sex",split=True,data=tips)
```

#### Types: 'scatter', 'reg', 'resid', 'kde', or 'hex'

```python
ax = sns.jointplot(x="total_bill",y="tip",data=tips)
```

```python
ax = sns.jointplot(x="total_bill",y="tip",kind="reg",data=tips)
```

```python
ax = sns.jointplot(x="total_bill",y="tip",kind="kde",data=tips)
```

#### Pairplot

```python
ax = sns.pairplot(tips,hue="sex")
```

#### LM plot

```python
ax = sns.lmplot(x="total_bill",y="tip",data=tips)
```

```python
ax = sns.lmplot(x="total_bill",y="tip",hue="smoker",data=tips)
```

Non smokers are paying higher tip as compared to smoker.

So the Rate of tip given by non-smokers are high.

*The padding is the 95% confidence interval for the set of points.*
