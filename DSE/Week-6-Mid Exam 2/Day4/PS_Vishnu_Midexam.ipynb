<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#a61652;">5A.Look at the year on year change in mean budget and mean revenue.Create a time series chart with mean revenue and budget per year?

```python
plot_it = df.groupby('year')[['budget','gross']].agg('mean')
print(plot_it)
plot_it.columns
```

```python
ax = sns.lineplot(x=plot_it.index,y=plot_it['budget'],label='budget')
sns.lineplot(x=plot_it.index,y=plot_it['gross'],label='gross')
plt.legend()
plt.show()
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#a61652;">1B)  Check the number of observations for "Musical" and create a new variable genre not equal to Musical ?(1 marks)

```python
musical = df['genre'] == "Musical"
# there are 4 rows which are musical in genre
print(sum(musical))

df['not_musical'] = ~musical
df['not_musical']
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#a61652;">1C) Use that new variable and plot the  Boxplot of revenue Vs genres?(2 marks)

```python
plt.figure(figsize=(8,12))
sns.boxplot(data=df,x="gross",y="genre",hue="not_musical")
plt.show()
```

As expected, only one genre is musical so that alone is show in a different colour.

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#a61652;"> 1D) Create a crosstab showing rating repartition by genre(1 marks)

```python
# My understanding from the question,
# To display the cross tabulation for rating v/s genre
# Showing the frequency per category
```

```python
rat_genre = pd.crosstab(index=df["rating"],columns=df["genre"],normalize="index")
rat_genre
```

<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#a61652;"> 1E) Create a stacked graph displaying the repartition of genre per content rating(1 marks)

```python
# I normalized the values to visualize
# proportion of genre per rating
rat_genre.plot.bar(stacked=True)
plt.show()
```
