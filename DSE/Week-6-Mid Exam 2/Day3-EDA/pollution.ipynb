```python
df = pd.read_csv('pollution.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
poll = df.groupby("date")[['so2','no2','rspm','spm']].agg('mean')
poll.reset_index(inplace=True)
```

```python
plt.plot(poll.index,poll['so2'])
```

```python
pollutants = ['so2','no2','rspm','spm']

for i,col in enumerate(pollutants):
    plt.figure(i)
    sns.lineplot(data=poll,x='date',y=col)
plt.show()
```

```python
pollutants = ['so2','no2','rspm','spm']

for i,col in enumerate(pollutants):
    plt.figure(i,figsize=(8,8))
    sns.lmplot(data=df,x='date',y=col)
plt.show()
```

```python
pollutants = ['so2','no2','rspm','spm']

for i,col in enumerate(pollutants):
    plt.figure(i)
    sns.lineplot(data=df,x='date',y=col,hue="type")
plt.show()
```
