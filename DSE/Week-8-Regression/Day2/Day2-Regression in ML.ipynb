```python
df = pd.read_csv('../Day1/data/airline cost new.csv')
```

```python
x = df['num_pass']
y = df['cost']
```

```python
m = 0 
c = 1000
l = 0.0001
epochs = 1000
n = len(x)

for i in range(epochs):
    y_pred = m*x + c
    D_m = (-2/n)*sum(x*(y - y_pred))
    D_c = (-2/n)*sum((y - y_pred))
    m = m - l*D_m
    c = c - l*D_c
    
    if i%100 == 0:
        plt.scatter(x,y)
        plt.plot(x,y_pred,color='r')
        plt.show()
print('m {}, c {}'.format(m,c))
```
