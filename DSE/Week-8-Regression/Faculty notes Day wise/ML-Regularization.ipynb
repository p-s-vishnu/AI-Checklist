# Regularization in Machine Learning

One of the major aspects of training your machine learning model is avoiding overfitting. The model will have a low accuracy if it is overfitting. This happens because your model is trying too hard to capture the noise in your training dataset. By noise we mean the data points that don’t really represent the true properties of your data, but random chance. Learning such data points, makes your model more flexible, at the risk of overfitting.

Regularization is a form of regression, that constrains or regularizes or shrinks the coefficient estimates towards zero. In other words, this technique discourages learning a more complex or flexible model, so as to avoid the risk of overfitting.

A simple linear regression looks like this:

y = $\beta$<sub>0</sub> + $\beta$<sub>1</sub>X<sub>1</sub> + $\beta$<sub>2</sub>X<sub>2</sub> + ... $\beta$<sub>j</sub>X<sub>j</sub>

Where,
<br>y = dependent variable
<br>X<sub>j</sub> = independent variable
<br>$\beta$<sub>j</sub> = coefficient estimates for the corresponding independent variable



The fitting procedure involves a loss function, known as residual sum of squares or RSS. The coefficients are chosen, such that they minimize this loss function.



Now, this will adjust the coefficients based on your training data. If there is noise in the training data, then the estimated coefficients won’t generalize well to the future data. This is where regularization comes in and shrinks or regularizes these learned estimates towards zero, since smaller coefficients generate simpler model.

Now, we add a regularization term to the loss function to keep the weights small, making the model simpler and avoiding overfitting. Now the coefficients are estimated by minimizing this function.

Here, $\lambda$ is regularization parameter (or tuning parameter) that decides how much we want to penalize the weights of our model. When $\lambda$ = 0, the penalty term has no eﬀect, and the estimates will be equal to our original values. However, as $\lambda$→∞, the impact of the shrinkage penalty grows, and the cofficient estimates will approach zero, this results in a very simple model having high bias, underfitting. As can be seen, selecting a good value of $\lambda$ is critical.

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression, Lasso, LassoCV, Ridge, RidgeCV, ElasticNet, ElasticNetCV
```

```python
# creating dependent and independent variables
X = pd.DataFrame(load_boston().data, columns=load_boston().feature_names)
y = load_boston().target

X.head()
```

```python
# predicting using simple linear regression model
model_lr = LinearRegression()
model_lr.fit(X,y)

coeff_result = pd.DataFrame(list(zip(X.columns, model_lr.coef_)), columns=["variable","lr_coefficient"])
model_result = pd.DataFrame(columns=["model","R-squared","RMSE","alpha"])
model_result = model_result.append(pd.DataFrame([["Linear Regression",
                 cross_val_score(model_lr, X, y, cv=3, scoring="r2", n_jobs=-1).mean(), 
                 cross_val_score(model_lr, X, y, cv=3, scoring="neg_mean_squared_error", n_jobs=-1).mean(),
                 np.NaN]], columns=["model","R-squared","RMSE","alpha"]))
```

## 1. Lasso regression



Lasso estimates the coefficients by minimizing the above function. In statistics, this is known as the L1 norm. It can be seen that lasso regression penalizes the high coefficients ($\beta$<sub>j</sub>). In the case of the lasso, the L1 penalty has the eﬀect of forcing some of the coeﬃcient estimates to be exactly equal to zero when the tuning parameter $\lambda$ is suﬃciently large. Therefore, the lasso method also performs variable selection.

```python
# hyperparameter tuning
list_alphas = list(np.linspace(0.1,5,1000))
model_ls = LassoCV(alphas=list_alphas, cv=3, n_jobs=-1)
model_ls.fit(X,y)
coeff_result = pd.concat([coeff_result, pd.DataFrame(model_ls.coef_, columns=["lasso_coefficient"])], axis=1, sort=False)
model_result = model_result.append(pd.DataFrame([["Lasso Regression",
                 cross_val_score(model_ls, X, y, cv=3, scoring="r2", n_jobs=-1).mean(), 
                 cross_val_score(model_ls, X, y, cv=3, scoring="neg_mean_squared_error", n_jobs=-1).mean(),
                 model_ls.alpha_]], columns=["model","R-squared","RMSE","alpha"]))
```

```python
# plotting mse path curve
plt.plot(model_ls.alphas_, model_ls.mse_path_.mean(1))
plt.axvline(model_ls.alpha_, color="g")
plt.xlabel("alpha")
plt.ylabel("Average MSE")
plt.title("Alpha vs MSE")
plt.show()
```

## 2. Ridge regression



Ridge estimates the coefficients by minimizing the above function. In statistics, this is known as the L2 norm. It can be seen that ridge regression penalizes the squares of coefficients ($\beta$<sub>j</sub><sup>2</sup>). It will shrink the coefficients for least important predictors, very close to zero. But it will never make them exactly zero. In other words, the final model will include all predictors. As $\lambda$ increases, ridge coefficient estimates will approach zero. 

```python
# hyperparameter tuning
list_alphas = list(np.linspace(0.01,1,100))
model_rg = RidgeCV(alphas=list_alphas, cv=3)
model_rg.fit(X,y)
coeff_result = pd.concat([coeff_result, pd.DataFrame(model_rg.coef_, columns=["ridge_coefficient"])], axis=1, sort=False)
model_result = model_result.append(pd.DataFrame([["Ridge Regression",
                 cross_val_score(model_rg, X, y, cv=3, scoring="r2").mean(), 
                 cross_val_score(model_rg, X, y, cv=3, scoring="neg_mean_squared_error").mean(),
                  model_rg.alpha_]], columns=["model","R-squared","RMSE","alpha"]))
```

## 3. Elastic Net regression



Elastic net estimates the coefficients by minimizing the above function. It is a combination of both lasso and ridge regression.

```python
# hyperparameter tuning
list_alphas = list(np.logspace(-2,0,1000))
model_en = ElasticNetCV(alphas=list_alphas, cv=3)
model_en.fit(X,y)
coeff_result = pd.concat([coeff_result, pd.DataFrame(model_en.coef_, columns=["elastic_net_coefficient"])], axis=1, sort=False)
model_result = model_result.append(pd.DataFrame([["ElasticNet Regression",
                 cross_val_score(model_en, X, y, cv=3, scoring="r2").mean(), 
                 cross_val_score(model_en, X, y, cv=3, scoring="neg_mean_squared_error").mean(),
                  model_en.alpha_]], columns=["model","R-squared","RMSE","alpha"]))
```

```python
# plotting mse path curve
plt.plot(model_en.alphas_, model_en.mse_path_.mean(1))
plt.axvline(model_en.alpha_, color="g")
plt.xlabel("alpha")
plt.ylabel("Average MSE")
plt.title("Alpha vs MSE")
plt.show()
```

## Summary

Regularization, significantly reduces the variance of the model, without substantial increase in its bias. So the tuning parameter λ, used in the regularization techniques described above, controls the impact on bias and variance. As the value of λ rises, it reduces the value of coefficients and thus reducing the variance. Till a point, this increase in λ is beneficial as it is only reducing the variance(hence avoiding overfitting), without loosing any important properties in the data. But after certain value, the model starts loosing important properties, giving rise to bias in the model and thus underfitting. Therefore, the value of λ should be carefully selected. It is a useful technique that can help in improving the accuracy of your regression models.
