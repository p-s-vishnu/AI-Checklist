### 10. Use the merged dataset to plot the following:  (3 + 4 marks)

    a) Generate boxplots for all the numerical variables in the dataset.
    b) Generate pairplots for the same dataset and point out the variables having a high correlation. 

````python
merged.info()
````

Numerical columns in the dataset are
- bookID (Irrelevant column)
- average_rating
- ratings_count

````python
def plot_box(col,title):
    merged.boxplot(column=col,figsize=(10,8))
    plt.title(title)
    plt.show()
````

### a) Boxplots

#### 1. Average rating

````python
plot_box("average_rating","Average ratings")
````

#### 2. Ratings count

````python
plot_box("ratings_count","Ratings count")
````

### b) Pairplots

````python
# Removing Book Id from the data frame because correlation with the column is not relevant
# Also removing the _merge which was created as part of the merge
````

````python
df_pair = merged.loc[:,"title" : "ratings_count"]
df_pair.columns
````

````python
sns.set()
sns.pairplot(df_pair,height=5)
plt.show()
````

We can see that, for pairplot, the columns under consideration are `average_rating` and `rating_count`.

As the `average_rating` increases `ratings_count` is not showing any linear changes (neither positive nor negative), hence we can conclude that these columns dont follow any correlation.

This can be confirmed by lookinh at the correlation matrix of the same.

````python
df_pair.corr()
````
