## Probabilistic Mass Function

```python
import scipy.stats as stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Binomial distribution

```python
# a
n= 10
p= 0.05

# Q1
x = 0
stats.binom.pmf(x,n,p)
```

59% there are no defects in the bulb

```python
# 2. Exactly one bulb

x = 1
stats.binom.pmf(x,n,p)
```

```python
# 3. Two or fewer bulbs are faulty
# x <= 2
x= 2

print(sum([stats.binom.pmf(i,n,p) for i in range(x+1)]))
# alternate method
print(stats.binom.cdf(x,n,p))
```

98% chance that 2 or fewer bulbs are faulty

```python
n,p,x = 10,0.42,np.arange(0,11)

binompmf = stats.binom.pmf(x,n,p)
binompmf
```

```python
plt.rcParams['figure.figsize'] = 5,5
plt.plot(binompmf)
```

```python
binomcdf = stats.binom.cdf(x,n,p)
plt.plot(binomcdf)
```

### Poisson distribution

```python
mu = 2.5
x = 0
stats.poisson.pmf(x,mu)
```

```python
1-stats.poisson.cdf(x,mu)
```

```python
mu = 3.4 # per minute
1 - stats.poisson.cdf(0,mu/2)
```

```python
plt.rcParams['figure.figsize'] = 9,5
x1 = np.arange(0,25)
pois = stats.poisson.pmf(x1,mu)
poiscdf = stats.poisson.cdf(x1,mu)
plt.plot(x1,pois,label="Probability dist function")
# plt.plot(x1,poiscdf,label="Cumulative dist function")
plt.xlabel('Number of customers(Discrete values)')
plt.ylabel('Discrete Probability')
plt.legend()
plt.show()
```

## Normal Distribution

```python
mu = 68
sigma = 12
x = 90
```

```python
# 1.How many users are spending time more than 90 minutes on messaging
1 - stats.norm.cdf(x,mu,sigma)
```

```python
# By computing z
z = (x-mu)/sigma
z
```

```python
# 2. What proportion of users are spending less than 20 minutes?
stats.norm.cdf(20,mu,sigma)
```

```python
x1 = 20

z1 = (x1 - mu)/sigma
stats.norm.cdf(z1)
```

```python
x4 = np.arange(0,114)
normp= stats.norm.pdf(x4,mu,sigma)
```

```python
plt.rcParams['figure.figsize'] = 8,4
plt.plot(normp)
plt.title('Normal Curve')
plt.xlabel('Time spent on messages')

plt.show()
```
